```python
# Import library
from transformers import ViltProcessor, ViltForQuestionAnswering
import requests
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Load image and question
url = "https://images.unsplash.com/photo-1728233363803-88a1226bf91d?w=600&auto=format&fit=crop&q=60&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxmZWF0dXJlZC1waG90b3MtZmVlZHw3NXx8fGVufDB8fHx8fA%3D%3D"
image = Image.open(requests.get(url, stream=True).raw)
question = "what is the skin color of the owl?"

# Load processor and model
processor = ViltProcessor.from_pretrained("dandelin/vilt-b32-finetuned-vqa")
model = ViltForQuestionAnswering.from_pretrained("dandelin/vilt-b32-finetuned-vqa")

# Prepare inputs
inputs = processor(image, question, return_tensors="pt")

# Forward pass
outputs = model(**inputs)
logits = outputs.logits

# Get predicted answer
predicted_idx = logits.argmax(dim=-1).item()
predicted_answer = model.config.id2label[predicted_idx]

# Print the predicted answer
# Display the image
plt.imshow(image)
plt.axis('off')  # Hide axis
plt.show()
print("Predicted answer:", predicted_answer)
```
